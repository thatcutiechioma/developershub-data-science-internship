```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(iris['petal_length'], kde=True)
plt.title("Distribution of Petal Length")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title("Sepal Length vs Width")
plt.show()
```

# Dataset Description

The Iris dataset contains 150 rows and 5 columns. The features include:

- sepal_length  
- sepal_width  
- petal_length  
- petal_width  
- species  

This dataset is commonly used for beginners in data science and machine learning. It contains three flower species: Iris-setosa, Iris-versicolor, and Iris-virginica.
